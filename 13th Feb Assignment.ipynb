Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
        
        ArithhmeticError class is the base class for those built-in exceptions that 
        are raised for various arithmetic errors such as: Overflow error, 
        ZeroDivisionError, FloatingPointError.
        The Arithmetic error occurs when an error is encountered during numeric calculations.
        

```python
# example of zeroDivision error 
try :
    a = 10/0
    print(a)
except ArithmeticError as e:
    print("This statement is raising an arithmetic error:", e)
else :
    print("success")
```

```python
# example of overflow error 
try:
    j = 5.0
    for i in range(1, 1000):
        j = j**i
        print(j)
except ArithmeticError as e:
    print("This statement is raising an arithmetic error:", e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

    The LookupError exception error in python forms the base class for all exceptions 
    that are raised when an index or a key is not found for a sequence or dictionary
    respectively. 

```python
# keyError example
try:
    a = {1:'a', 2:'b', 3:'c'}
    print(a[4])
except LookupError as e:
    print("Key Error Exception raised:", e)
else :
    print("Success, no error!")
```

```python
# Index Error
try: 
    l = [1,2,4]
    print(l[4])
except LookupError as e:
    print("Index Error Exception raised:", e)
```
